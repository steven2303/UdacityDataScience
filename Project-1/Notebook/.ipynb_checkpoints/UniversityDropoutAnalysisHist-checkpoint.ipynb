## ¿Cómo se correlaciona el desempeño académico de los estudiantes, como las calificaciones y las tasas de aprobación, con su situación financiera, como los pagos tardíos o la falta de pago?

```python
saldos = saldos_data[(saldos_data.term < 201902) & (saldos_data.term > 201700)].copy()
semanal = semanal_data[(semanal_data.term < 201902) & (semanal_data.term > 201700)].copy()
```

```python
semestral = semestral_data[(semestral_data.term < 201902) & (semestral_data.term > 201700)].copy()
```

```python
saldos.payment1dl = pd.to_datetime(saldos.payment1dl)
saldos.payment2dl = pd.to_datetime(saldos.payment2dl)
saldos.payment3dl = pd.to_datetime(saldos.payment3dl)
saldos.payment4dl = pd.to_datetime(saldos.payment4dl)
saldos.payment5dl = pd.to_datetime(saldos.payment5dl)
```

```python
saldos.payment1date = pd.to_datetime(saldos.payment1date)
saldos.payment2date = pd.to_datetime(saldos.payment2date)
saldos.payment3date = pd.to_datetime(saldos.payment3date)
saldos.payment4date = pd.to_datetime(saldos.payment4date)
saldos.payment5date = pd.to_datetime(saldos.payment5date)
```

```python
df_max_week = semanal.groupby('term').week.max().reset_index().rename(columns = {'week':'max_week'})
semanal_ = semanal.merge(df_max_week,how = 'left',on = 'term')
semanal_ = semanal_[semanal_.week == semanal_.max_week]
```

```python
cut_semanal = semanal_[['student','term','classenddate']].drop_duplicates(subset = ['student','term'])
saldos = saldos.merge(cut_semanal ,how = 'inner', on = ['student','term'])
```

Contamos con la fecha de termino de cada uno de los diferentes periodos academicos de la universidad, compararemos la fecha de pago de cada boleta con la fecha de termino del periodo, y setearemos a 0 en caso la fecha sea mayor.

```python
saldos.classenddate = pd.to_datetime(saldos.classenddate)
```

```python
saldos.loc[saldos['payment1date'] > saldos['classenddate'],'payment1'] = 0
saldos.loc[saldos['payment2date'] > saldos['classenddate'],'payment2'] = 0
saldos.loc[saldos['payment3date'] > saldos['classenddate'],'payment3'] = 0
saldos.loc[saldos['payment4date'] > saldos['classenddate'],'payment4'] = 0
saldos.loc[saldos['payment5date'] > saldos['classenddate'],'payment5'] = 0
```

```python
saldos[['payment1','payment2','payment3','payment4','payment5']] = saldos[[
            'payment1','payment2','payment3','payment4','payment5']].fillna(0)
```

```python
saldos.loc[saldos['payment1'] == saldos['payment1amount'],'pago1completo'] = 1
saldos.loc[saldos['payment2'] == saldos['payment2amount'],'pago2completo'] = 1
saldos.loc[saldos['payment3'] == saldos['payment3amount'],'pago3completo'] = 1
saldos.loc[saldos['payment4'] == saldos['payment4amount'],'pago4completo'] = 1
saldos.loc[saldos['payment5'] == saldos['payment5amount'],'pago5completo'] = 1
saldos[['pago1completo','pago2completo','pago3completo','pago4completo','pago5completo']] = saldos[[
            'pago1completo','pago2completo','pago3completo','pago4completo','pago5completo']].fillna(0)

saldos.loc[saldos['payment1'] == 0,'pago1'] = 0
saldos.loc[saldos['payment2'] == 0,'pago2'] = 0
saldos.loc[saldos['payment3'] == 0,'pago3'] = 0
saldos.loc[saldos['payment4'] == 0,'pago4'] = 0
saldos.loc[saldos['payment5'] == 0,'pago5'] = 0
saldos[['pago1','pago2','pago3','pago4','pago5']] = saldos[[
            'pago1','pago2','pago3','pago4','pago5']].fillna(1)
```

```python
saldos['pct_pagado'] = round((saldos['payment1'] + saldos['payment2'] + saldos['payment3'] + saldos['payment4'] + 
                    saldos['payment5'])/( saldos['payment1amount'] + saldos['payment2amount'] + saldos['payment3amount'] +
                                        saldos['payment4amount'] + saldos['payment5amount'] +1),3)
```
